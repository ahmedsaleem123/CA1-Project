**Dataset Columns:**Here's a brief overview of the columns in our cleaned dataset:<br><br>

**1.Basic Information:**

 - **continent:** Categorization by continent.
 - **location:** Specific location or country.
 - **date:** Date of observation.<br><br>

**2. COVID-19 Metrics:**

 - **total_cases:** Total reported COVID-19 cases.
 - **new_cases_smoothed:** Smoothed daily count of new cases.
 - **total_deaths:** Total reported COVID-19 deaths.
 - **new_deaths_smoothed:** Smoothed daily count of new deaths.
 - **total_vaccinations:** Total number of COVID-19 vaccinations administered.
 - **people_fully_vaccinated:** Number of people fully vaccinated.
 - **new_people_vaccinated_smoothed:** Smoothed daily count of new people vaccinated.<br><br>

**3. Demographic & Socioeconomic Factors:**

 - **median_age:** Median age of the population.
 - **gdp_per_capita:** GDP per capita (economic measure).
 - **diabetes_prevalence:** Prevalence of diabetes in the population.
 - **female_smokers:** Percentage of female smokers.
 - **male_smokers:** Percentage of male smokers.
 - **population:** Total population count.
