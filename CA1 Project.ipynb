```python
# Identifying numerical variables
num_df = data.select_dtypes(include=['number'])
num_cols = num_df.columns
print("The number of numerical columns in our dataset is: " + str(len(num_cols)))
print("The numerical columns in our dataset are: ", num_cols)
```

```python
# Identifying Discrete and Continuous Variables

# Select numerical columns
numerical_columns = data.select_dtypes(include=['number']).columns

# Categorize variables based on fractional parts, ignoring null values
discrete_vars = []
continuous_vars = []

for col in numerical_columns:
    non_null_values = data[col].dropna()  # Ignore null values in the column

    if (non_null_values % 1 == 0).all():  # Checking if all non-null values have zero fractional parts
        discrete_vars.append(col)
    if not (non_null_values % 1 == 0).all():  # Checking if any non-null value has a non-zero fractional part
        continuous_vars.append(col)

# Print identified variables
print("Discrete Variables:", discrete_vars)
print("Continuous Variables:", continuous_vars)
```

```python
# Discrete Variable: total_cases

# Convert 'date' column to datetime if it's not already in datetime format
data['date'] = pd.to_datetime(data['date'])

# Filter data for December 2021
dec_2021_data = data[(data['date'].dt.year == 2021) & (data['date'].dt.month == 12)]

# Group by 'continent' and 'location' to calculate total cases for each continent in December 2021
continent_cases_dec_2021 = dec_2021_data.groupby(['continent', 'location'])['total_cases'].max().reset_index()

# Aggregate total cases for each continent
continent_total_cases = continent_cases_dec_2021.groupby('continent')['total_cases'].sum().reset_index()

# Create a bar plot for total cases in each continent
plt.figure(figsize=(10, 6))
plt.bar(continent_total_cases['continent'], continent_total_cases['total_cases'], color='skyblue')
plt.xlabel('Continent')
plt.ylabel('Total Cases')
plt.title('Total COVID-19 Cases in Each Continent in December 2021')
plt.xticks(rotation=90)
plt.tight_layout()
plt.show()

```

```python
# Discrete Variable: total_deaths

# Filter data for Asian countries in June 2022
asian_countries_june_2022 = data[(data['continent'] == 'Asia') &
                                 (data['date'].dt.year == 2022) &
                                 (data['date'].dt.month == 6)]

# Plotting total_deaths in Asian countries for June 2022
plt.figure(figsize=(16, 6))
plt.bar(asian_countries_june_2022['location'], asian_countries_june_2022['total_deaths'], color='salmon')
plt.xlabel('Asian Countries')
plt.ylabel('Total Deaths')
plt.title('Total Deaths in Asian Countries - June 2022')
plt.xticks(rotation=90)
plt.tight_layout()

plt.show()
```

```python
# Discrete Variable: new_cases

# Filter the data for August 2020 and European countries
aug_2020_data = data[(data['date'].dt.year == 2020) & (data['date'].dt.month == 8) & (data['continent'] == 'Europe')]

# Creating Bar Plot for New Cases in August 2020 for European Countries
plt.figure(figsize=(16, 6))
plt.bar(aug_2020_data['location'], aug_2020_data['new_cases'], color='red')
plt.xlabel('European Countries')
plt.ylabel('Number of New Cases')
plt.title('New COVID-19 Cases in August 2020 (European Countries)')
plt.xticks(rotation=90)
plt.tight_layout()

plt.show()
```

```python
# Continuous Variable: total_cases_per_million

# Filter data for African countries
africa_data = data[(data['continent'] == 'Africa') & (~data['total_cases_per_million'].isnull())]

# Create an interactive scatter plot using Plotly Express
fig = px.scatter(africa_data, x='location', y='total_cases_per_million',
                 color='location', title='Total COVID-19 Cases per Million in African Countries',
                 labels={'location': 'Country', 'total_cases_per_million': 'Total Cases per Million'})

# Update layout for better readability
fig.update_layout(xaxis={'categoryorder':'total ascending'},
                  xaxis_title='Country', yaxis_title='Total Cases per Million',
                  height=600, width=1000)

# Show the interactive plot
fig.show()
```

```python
# # Continuous Variable: new_cases_smoothed

# Filter data for South American countries and 'new_cases_smoothed'
south_america_data = data[data['continent'] == 'South America']

# Create an interactive line plot using Plotly Express
line_fig = px.line(south_america_data,
                    x='location', y='new_cases_smoothed',
                    color='location',
                    title='New Cases Smoothed in South American Countries (Line Plot)',
                    labels={'new_cases_smoothed': 'New Cases Smoothed'})

line_fig.update_traces(mode='lines+markers')

line_fig.update_layout(xaxis_title='Country', yaxis_title='New Cases Smoothed')

line_fig.show()
```

```python
# Continuous Variable: new_cases_per_million

# Filter data for Asian countries and January 2023
jan_2023_asia_data = data[(data['continent'] == 'Asia') & (data['date'].dt.year == 2023) & (data['date'].dt.month == 1)]

# Create an interactive choropleth map using Plotly Express
fig = px.choropleth(jan_2023_asia_data,
                    locations="iso_code",  # Assuming 'iso_code' is the ISO code column for countries
                    color="new_cases_per_million",
                    hover_name="location",
                    projection="natural earth",  # Adjust projection as needed
                    title='New Cases per Million in Asian Countries (January 2023)',
                    labels={'new_cases_per_million': 'New Cases per Million'})

fig.update_geos(showcountries=True, countrycolor="Black")  # Add country borders
fig.update_layout(coloraxis_colorbar=dict(title='New Cases per Million'))  # Update colorbar title

fig.show()
```
